```python
from yahoo_fin import options
import pandas as pd
import requests
from io import StringIO
```

```python
# Define the stock ticker symbol
stock = 'AAPL'

# Add headers to the request
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36'
}

# Fetch the options chain page
url = f"https://finance.yahoo.com/quote/{stock}/options?p={stock}"
response = requests.get(url, headers=headers).text

# Wrap the response text with StringIO
try:
    tables = pd.read_html(StringIO(response))
    options_chain = tables[0]  # Typically the first table contains the data
    print(options_chain)
except ValueError:
    print("Could not find any tables. The response might be blocked by Yahoo.")
```
